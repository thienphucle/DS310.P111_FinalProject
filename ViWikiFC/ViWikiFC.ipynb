```python
# Tỉ lệ claim có evidence
def calculate_claim_with_evidence_ratio(data):
    return data['evidence'].apply(lambda x: 1 if str(x).strip() else 0).mean() * 100

print("Tỉ lệ claim có evidence:")
print(f"Train set: {calculate_claim_with_evidence_ratio(train_data):.2f}%")
print(f"Dev set: {calculate_claim_with_evidence_ratio(dev_data):.2f}%")
print(f"Test set: {calculate_claim_with_evidence_ratio(test_data):.2f}%\n")
```

```python
import re
# Tiền xử lý dữ liệu cho ViFactCheck
def clean_text(text):
    text = str(text).lower()  # Chuyển về chữ thường
    text = re.sub(r'\s+', ' ', text)  # Loại bỏ khoảng trắng thừa
    text = re.sub(r'[^\w\s]', '', text)  # Loại bỏ dấu câu
    return text.strip()

def tokenize_text(text):
    return ' '.join(text.split())  # Tokenize đơn giản

# Tiền xử lý dữ liệu cho ViFactCheck
vifact_df['clean_statement'] = vifact_df['Statement'].apply(clean_text).apply(tokenize_text)
vifact_df['clean_evidence'] = vifact_df['Evidence'].apply(clean_text).apply(tokenize_text)

# Hiển thị vài dòng đầu của dữ liệu đã xử lý
vifact_df.head()
```

```python
# 6. So sánh hai bộ
def analyze_file(data, name):
    num_samples = len(data)
    avg_claim_length = data['claim'].apply(lambda x: len(str(x).split())).mean()
    avg_evidence_length = data['evidence'].apply(lambda x: len(str(x).split())).mean()
    label_distribution = data['gold_label'].value_counts(normalize=True)
    has_evidence_ratio = data['evidence'].apply(lambda x: 1 if str(x).strip() else 0).mean()
    
    print(f"--- {name} Analysis ---")
    print(f"Number of samples: {num_samples}")
    print(f"Avg. claim length: {avg_claim_length:.2f} words")
    print(f"Avg. evidence length: {avg_evidence_length:.2f} words")
    print(f"Label distribution:\n{label_distribution}")
    print(f"Claims with evidence: {has_evidence_ratio * 100:.2f}%")
    print("\n")
    
    return {
        'num_samples': num_samples,
        'avg_claim_length': avg_claim_length,
        'avg_evidence_length': avg_evidence_length,
        'label_distribution': label_distribution,
        'has_evidence_ratio': has_evidence_ratio
    }
```
