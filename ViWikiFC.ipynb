```python
# 1. Kích thước tệp (số lượng mẫu trong train, dev, test)
print("Kích thước từng tập:")
print(f"Train set: {train_data.shape[0]} samples")
print(f"Dev set: {dev_data.shape[0]} samples")
print(f"Test set: {test_data.shape[0]} samples\n")
```

```python
# 4. Tỉ lệ claim có evidence
def calculate_claim_with_evidence_ratio(data):
    return data['evidence'].apply(lambda x: 1 if str(x).strip() else 0).mean() * 100

print("Tỉ lệ claim có evidence:")
print(f"Train set: {calculate_claim_with_evidence_ratio(train_data):.2f}%")
print(f"Dev set: {calculate_claim_with_evidence_ratio(dev_data):.2f}%")
print(f"Test set: {calculate_claim_with_evidence_ratio(test_data):.2f}%\n")
```

```python
# 2. Độ dài trung bình của claim và evidence
def calculate_avg_length(data, column):
    return data[column].dropna().apply(lambda x: len(str(x).split())).mean()

print("Độ dài trung bình của claim và evidence:")
print(f"Train set: Avg. claim length = {calculate_avg_length(train_data, 'claim'):.2f} words, "
      f"Avg. evidence length = {calculate_avg_length(train_data, 'evidence'):.2f} words")
print(f"Dev set: Avg. claim length = {calculate_avg_length(dev_data, 'claim'):.2f} words, "
      f"Avg. evidence length = {calculate_avg_length(dev_data, 'evidence'):.2f} words")
print(f"Test set: Avg. claim length = {calculate_avg_length(test_data, 'claim'):.2f} words, "
      f"Avg. evidence length = {calculate_avg_length(test_data, 'evidence'):.2f} words\n")
```

```python
# 3. Phân phối nhãn (label) trong từng tệp
def plot_label_distribution(data, dataset_name):
    plt.figure(figsize=(8, 6))
    sns.countplot(x='gold_label', data=data, order=data['gold_label'].value_counts().index, palette="Set2")
    plt.title(f"Label Distribution in {dataset_name}")
    plt.xlabel("Label")
    plt.ylabel("Count")
    plt.show()

print("Phân phối nhãn trong từng tập:")
plot_label_distribution(train_data, "Train Set")
plot_label_distribution(dev_data, "Dev Set")
plot_label_distribution(test_data, "Test Set")
```

```python
# 5. Phân tích đặc điểm của dữ liệu
print("Phân tích tổng quan đặc điểm của dữ liệu ViWikiFC:")
print(f"- Tổng số mẫu: {data.shape[0]}")
print(f"- Số lượng feature: {data.shape[1]}")
print(f"- Số lượng claim duy nhất: {data['claim'].nunique()}")
print(f"- Số lượng topic (title) duy nhất: {data['title'].nunique()}")
print(f"- Claim dài nhất: {data['claim'].apply(lambda x: len(str(x).split())).max()} từ")
print(f"- Evidence dài nhất: {data['evidence'].apply(lambda x: len(str(x).split())).max()} từ\n")
```

```python
# 6. So sánh hai bộ
def analyze_file(data, name):
    num_samples = len(data)
    avg_claim_length = data['claim'].apply(lambda x: len(str(x).split())).mean()
    avg_evidence_length = data['evidence'].apply(lambda x: len(str(x).split())).mean()
    label_distribution = data['gold_label'].value_counts(normalize=True)
    has_evidence_ratio = data['evidence'].apply(lambda x: 1 if str(x).strip() else 0).mean()
    
    print(f"--- {name} Analysis ---")
    print(f"Number of samples: {num_samples}")
    print(f"Avg. claim length: {avg_claim_length:.2f} words")
    print(f"Avg. evidence length: {avg_evidence_length:.2f} words")
    print(f"Label distribution:\n{label_distribution}")
    print(f"Claims with evidence: {has_evidence_ratio * 100:.2f}%")
    print("\n")
    
    return {
        'num_samples': num_samples,
        'avg_claim_length': avg_claim_length,
        'avg_evidence_length': avg_evidence_length,
        'label_distribution': label_distribution,
        'has_evidence_ratio': has_evidence_ratio
    }
```

```python
viwikifc_train_analysis = analyze_file(train_data, "ViWikiFC Train")
viwikifc_dev_analysis = analyze_file(dev_data, "ViWikiFC Dev")
viwikifc_test_analysis = analyze_file(test_data, "ViWikiFC Test")
```

```python
viwikifc_results = {
    'train': viwikifc_train_analysis,
    'dev': viwikifc_dev_analysis,
    'test': viwikifc_test_analysis
}
```

```python
vifactcheck_train = pd.read_csv('C:\\Users\\ngmai\\Downloads\\ViFactCheck\\train_vifactcheck.csv')
vifactcheck_dev = pd.read_csv('C:\\Users\\ngmai\\Downloads\\ViFactCheck\\dev_vifactcheck.csv')
vifactcheck_test = pd.read_csv('C:\\Users\\ngmai\\Downloads\\ViFactCheck\\test_vifactcheck.csv')
```

```python
vifactcheck_train_analysis = analyze_file(vifactcheck_train, "ViFactCheck Train")
vifactcheck_dev_analysis = analyze_file(vifactcheck_dev, "ViFactCheck Dev")
vifactcheck_test_analysis = analyze_file(vifactcheck_test, "ViFactCheck Test")
```

```python
vifactcheck_results = {
    'train': vifactcheck_train_analysis,
    'dev': vifactcheck_dev_analysis,
    'test': vifactcheck_test_analysis
}
```

```python
# So sánh ViWikiFC và ViFactCheck
def compare_results(results1, results2, name1, name2):
    comparison = pd.DataFrame({
        'Metric': [
            'Avg. claim length', 'Avg. evidence length', 
            'Claims with evidence (%)', 'Number of samples'
        ],
        name1: [
            results1['train']['avg_claim_length'],
            results1['train']['avg_evidence_length'],
            results1['train']['has_evidence_ratio'] * 100,
            results1['train']['num_samples']
        ],
        name2: [
            results2['train']['avg_claim_length'],
            results2['train']['avg_evidence_length'],
            results2['train']['has_evidence_ratio'] * 100,
            results2['train']['num_samples']
        ]
    })
    
    print("\n--- Comparison Between Datasets ---")
    print(comparison)
    
    # Visualization
    comparison.set_index('Metric').plot(kind='bar', figsize=(12, 6), color=['#1f77b4', '#ff7f0e'])
    plt.title(f"Comparison of {name1} and {name2}")
    plt.ylabel("Value")
    plt.xticks(rotation=45)
    plt.show()
```

```python
compare_results(viwikifc_results, vifactcheck_results, "ViWikiFC", "ViFactCheck")
```
